```python
ax=sns.countplot(data=df,x="Education",hue='sex')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
emp_details=df.groupby(['Education','Gender']).size().reset_index(name='Count')
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=emp_details,x='Education',y='Count',hue='Gender')
print(emp_details)
```

```python
emp_details=df.groupby(['Age','Gender']).size().reset_index(name='Count')
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=emp_details,x='Age',y='Count',hue='Gender')
print(emp_details)
```

```python
emp_details=df.groupby(['City','Gender']).size().reset_index(name='Count')
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=emp_details,x='City',y='Count',hue='Gender')
print(emp_details)
```
